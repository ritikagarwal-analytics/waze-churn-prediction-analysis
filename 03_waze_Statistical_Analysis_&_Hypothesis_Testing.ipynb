To better understand behavioral differences between user segments, a statistical hypothesis test was conducted on the Waze dataset.

# **Data exploration and hypothesis testing**

### **Imports and data loading**

```python
# Import any relevant packages or libraries
import pandas as pd
from scipy import stats
```

```python
# Load dataset into dataframe
df = pd.read_csv('waze_dataset.csv')
```

## Descriptive Statistical Analysis

Descriptive statistics were used to summarize and compare key behavioral variables across user segments. 

In particular, summary measures such as mean and median values provided a quick comparison of average driving activity across device types. This helped identify preliminary differences between groups before conducting formal hypothesis testing.

### **Data exploration**

### Encoding Categorical Variable

The `device` variable is categorical (`iPhone` and `Android`). 

To enable statistical testing, the variable was encoded numerically by creating a new feature `device_type`. A mapping strategy was applied to convert device labels into integer representations while preserving the original column.

```python
# 1. Create `map_dictionary`
map_dictionary = {'Android': 2, 'iPhone': 1}

# 2. Create new `device_type` column
df['device_type'] = df['device']

# 3. Map the new column to the dictionary
df['device_type'] = df['device_type'].map(map_dictionary)

df['device_type'].head()
```

You are interested in the relationship between device type and the number of drives. One approach is to look at the average number of drives for each device type. Calculate these averages.

```python
df.groupby('device_type')['drives'].mean()
```

Preliminary descriptive analysis suggests that users interacting with the application via iPhone exhibit a higher average number of drives compared to Android users.

However, observed differences in sample means may result from random variation rather than a true underlying behavioral difference. To formally evaluate whether this difference is statistically significant, a two-sample hypothesis test was conducted.


### **Hypothesis testing**

Recall the difference between the null hypothesis ($H_0$) and the alternative hypothesis ($H_A$).

**Question:** What are your hypotheses for this data project?


**Hypotheses:**

$H_0$: There is no difference in average number of drives between drivers who use iPhone devices and drivers who use Androids.

$H_A$: There is a difference in average number of drives between drivers who use iPhone devices and drivers who use Androids.

A two-sample t-test was conducted at a 5% significance level to determine whether a statistically significant difference exists in the average number of drives between iPhone and Android users.

Since equality of variances cannot be assumed with certainty, Welch’s t-test (unequal variance assumption) was used to provide a more robust comparison between the two groups.

```python
# 1. Isolate the `drives` column for iPhone users.
iPhone = df[df['device_type'] == 1]['drives']

# 2. Isolate the `drives` column for Android users.
Android = df[df['device_type'] == 2]['drives']

# 3. Perform the t-test
stats.ttest_ind(a=iPhone, b=Android, equal_var=False)
```

 Based on the p-value you got above, do you reject or fail to reject the null hypothesis?

> *Since the p-value is larger than the chosen significance level (5%), you fail to reject the null hypothesis. You conclude that there is **not** a statistically significant difference in the average number of drives between drivers who use iPhones and drivers who use Androids.*

### **Communicate insights with stakeholders**

## Business Interpretation

The hypothesis test indicates that there is no statistically significant difference in the average number of drives between iPhone and Android users.

From a business perspective, this suggests that device type alone is unlikely to be a primary driver of user engagement or churn risk.

Future analysis should focus on behavioral and usage-based variables that may better explain differences in driving activity. Additional hypothesis testing and targeted experiments could help identify stronger predictors of user retention.
